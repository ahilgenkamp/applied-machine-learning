```python
import pandas as pd
import numpy as np
import os
import time
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

#matplot lib notebooks display command
%matplotlib inline

#SET OPTIONS
pd.options.display.float_format = '{:,.2f}'.format
#GLOBAL VARIABLES
cwd = os.getcwd()
```

```python
#load dataset
result_df = pd.read_csv(cwd+'\\non_automated_invoices_with_acct_info_sample.csv', encoding='utf-8')
result_df.describe()
```

```python
plt.figure(figsize=(18,5))
plt.boxplot(result_df['DAYS_TO_PAY'], vert=False, showfliers=True, showmeans=True, labels=['DAYS_TO_PAY'])
plt.show()
```

```python
sns.catplot(x='BILLING_LANGUAGE', y='DAYS_TO_PAY', kind="box", data=result_df, height=8.27, aspect=12.5/8.27)
plt.xticks(rotation=45)
plt.show()
```

```python
sns.catplot(x='CR_CHK_CD', y='DAYS_TO_PAY', kind="box", data=result_df, height=8.27, aspect=14.5/8.27)
plt.xticks(rotation=90)
plt.show()
```

```python
sns.jointplot(result_df['USAGE_SCORE'], result_df['DAYS_TO_PAY'], kind='reg'
              , stat_func=stats.pearsonr, joint_kws={'line_kws':{'color':'orange'}}, scatter_kws={"s": 1}) 
plt.show()
```
